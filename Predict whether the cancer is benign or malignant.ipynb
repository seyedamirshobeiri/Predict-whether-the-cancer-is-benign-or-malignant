```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
import sklearn.metrics as met
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('data.csv')
```

```python
data['diagnosis'].value_counts()
```

```python
data.diagnosis=data.diagnosis.apply(['B','M'].index)
```

```python
data
```

```python
label=data['diagnosis']
data=data.drop(['diagnosis','Unnamed: 32','id'],axis=1)
```

```python
data.info()
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,label,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
clf=MLPClassifier(max_iter=200,verbose=True)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)

X_train = scaler.transform(x_train)
X_test = scaler.transform(x_test)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
pre=clf.predict(X_test)
met.f1_score(y_test,pre)
```

```python
accuracy=met.accuracy_score(y_test,pre)
```

```python
print('The accuracy of model is : ',int(accuracy*100),'%')
```
